Que 1: Seaborn is a data visualization library built on top of Matplotlib, and it provides a high-level interface for creating attractive and informative statistical graphics. Here are five common plots that you can create using Seaborn along with their uses:

1) Scatter Plot (sns.scatterplot()): Scatter plots are used to visualize the relationship between two continuous variables.   They are helpful for identifying patterns, clusters, or correlations in data.

2) Line Plot (sns.lineplot()): Line plots are used to visualize trends and changes in data over a continuous axis, typically time. They are suitable for showing how a variable evolves.

3) Box Plot (sns.boxplot()): Box plots are used to display the distribution of a continuous variable, including measures like quartiles and potential outliers. They are useful for identifying skewness and spread.

4) Bar Plot (sns.barplot()): Bar plots are used to compare the mean or aggregate values of different categories. They are suitable for visualizing group-wise comparisons.

5) Pair Plot (sns.pairplot()): Pair plots are used to visualize pairwise relationships between multiple variables in a dataset. They are helpful for exploring correlations and distributions simultaneously.



```python
#Que 2.

import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Create a line plot for different events and regions
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)

plt.title("FMRI Signal Over Time")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.show()
```

```python
#Que3.
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Create two box plots for 'pclass' vs. 'age' and 'pclass' vs. 'fare'
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.title('Box Plot of Age by Pclass')

plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.title('Box Plot of Fare by Pclass')

plt.tight_layout()
plt.show()
```

```python
#Que4.
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Create a histogram with 'price' and 'cut' as hue
sns.histplot(data=diamonds_data, x="price", hue="cut", bins=30, kde=True)
plt.title("Histogram of Diamond Prices by Cut")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
#Que5.
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset
iris_data = sns.load_dataset("iris")

# Create a pair plot with hue as "species"
sns.pairplot(iris_data, hue="species", markers=["o", "s", "D"])
plt.title("Pair Plot of Iris Dataset")
plt.show()
```

```python
#Que6.
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix for the heatmap
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(flights_pivot, cmap='YlGnBu', annot=True, fmt='d', linewidths=.5)
plt.title("Flights Data Heatmap")
plt.show()
```
