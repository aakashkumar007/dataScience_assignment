Q1: Matplotlib is a popular Python library used for data visualization. It provides a wide range of functions and tools for creating various types of plots and charts. Matplotlib is often used to visualize data to better understand patterns, trends, and relationships within the data .Some common plots that can be created using the Pyplot module of Matplotlib include:

Line plots: Used to represent data points connected by lines, often used for time series data.

Scatter plots: Display individual data points as markers, helpful for visualizing the relationship between two variables.

Bar plots: Show data as rectangular bars, useful for comparing categories or groups.

Histograms: Illustrate the distribution of a single variable, typically used for continuous data.

Pie charts: Display data as slices of a circle, showing the proportions of different categories in a whole.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

Q3: The subplot() function in Matplotlib is used to create multiple plots (subplots) within a single figure. It allows you to arrange multiple plots in a grid or other configurations. You specify the number of rows and columns in the grid and the position of each subplot within that grid.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])

# Creating the subplots
plt.subplot(2, 2, 1)
plt.plot(x, np.array([0, 100, 200, 300, 400, 500]))
plt.title("Line 1")

plt.subplot(2, 2, 2)
plt.plot(x, np.array([50, 20, 40, 20, 60, 70]))
plt.title("Line 2")

plt.subplot(2, 2, 3)
plt.plot(x, np.array([10, 20, 30, 40, 50, 60]))
plt.title("Line 3")

plt.subplot(2, 2, 4)
plt.plot(x, np.array([200, 350, 250, 550, 450, 150]))
plt.title("Line 4")

plt.show()
```

Q4: A bar plot, also known as a bar chart or bar graph, is used to display categorical data using rectangular bars of varying heights. It is often used to compare and visualize the quantities or frequencies of different categories or groups. A horizontal bar plot is essentially the same as a bar plot, but with the bars arranged horizontally.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Vertical bar plot
plt.figure(figsize=(8, 4))
plt.bar(company, profit, color='skyblue')
plt.title("Company Profits (Vertical Bar Plot)")
plt.xlabel("Company")
plt.ylabel("Profit (in millions)")
plt.show()

# Horizontal bar plot
plt.figure(figsize=(8, 4))
plt.barh(company, profit, color='lightgreen')
plt.title("Company Profits (Horizontal Bar Plot)")
plt.xlabel("Profit (in millions)")
plt.ylabel("Company")
plt.show()
```

Q5: A box plot, also known as a box-and-whisker plot, is used to display the distribution and summary statistics of a dataset, including measures such as the median, quartiles, and potential outliers. It is particularly useful for visualizing the spread and skewness of the data and for identifying any extreme values.



```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data)
plt.title("Box Plot")
plt.xlabel("Box")
plt.ylabel("Value")
plt.xticks([1, 2], ["Box 1", "Box 2"])
plt.show()
```
